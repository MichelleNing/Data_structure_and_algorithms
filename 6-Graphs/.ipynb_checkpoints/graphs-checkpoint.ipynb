## Connectivity

A connected graph with no unconnected nodes. 

<img src='connection.png' width=500>

There are two graphs, considered they described two different groups of friendships. The one on the right shows stronger relationship. It has less people, but with more connectoins. On the left group, once one connection is removed, the whole group could be destroyed. 

Connectivity measure the minimum number of elements that need to be removed to form a graph to be disconnected. It can be used to check which relationship is stronger. 

**Disconnected**

Disconnected graphs are very similar whether the graph's directed or undirectedâ€”there is some vertex or group of vertices that have no connection with the rest of the graph.

**Weakly Connected**

A directed graph is weakly connected when only replacing all of the directed edges with undirected edges can cause it to be connected. Imagine that your graph has several vertices with one outbound edge, meaning an edge that points from it to some other vertex in the graph. There's no way to reach all of those vertices from any other vertex in the graph, but if those edges were changed to be undirected all vertices would be easily accessible. 

**Connected**

Here we only use "connected graph" to refer to undirected graphs. In a connected graph, there is some path between one vertex and every other vertex.

**Strongly Connected**

Strongly connected directed graphs must have a path from every node and every other node. So, there must be a path from A to B AND B to A.

## Graph Representations

If objects are used to describe, can create vertex object and edge object. And give them each properties like name, strength, id number.

**Edge List**
<img src='edge-list.png' width=400>
Edge itself can be represent by a list with two elements.Those elements are normally numbers corresponding to the id numbers for vertexes. 

Edge List sometimes is called 2-dimension list or 2D list.

**Adjacency List**
<img src='adjacency-list.png' width=400>
In adjacency list, normally vertexes have id number that correspondences to the index in the right. 

In the array, each space is used to store a list of nodes. The index 0 represents the vertex with ID 0. That vertex edge with one node, so the reference to that node is stored in the first spot in the array.

Adjacency list is also a 2D list, and the list can contain lists.

## Adjacency Maxtrix

<img src='adjacency-matrix.png' width=400>
It is similar to the adjacency list. The indexes of the array still represent the node IDs, and the lists inside still represent which nodes are adjacent. However, these information in the list are slid in a different way. If there is a edge between two nodes, 1 goes into the array. If no edge, 0 goes into the array. A single edge shows twice in the array.
